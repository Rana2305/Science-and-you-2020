```python
# remove special character
data.columns=data.columns.str.replace('_','')
print(data.columns)
```

```python
#Items 22: A l‚Äôoccasion de la crise du coronavirus, on a beaucoup parl√© des ¬´ fakes news ¬ª ou des infos des fausses informations. 
#Dans quelle mesure faites-vous confiance aux personnes suivantes pour vous dire la v√©rit√© sur le coronavirus ? 
Items22=['ScientifiquesUniversitairesConfiance','ScientifiquesIndustrieConfiance','IngenieursConfiance'
               ,'PoliticiensConfiance','JuristesConfiance','journalistesConfiance','IntellectuelsConfiance'
               ,'LeadersReligieuxConfiance','OMSConfiance','MinistreSanteConfiance','MedecinConfiance']
```

```python
for i in data[Items22].columns:
    data[i] = pd.Categorical(data[i], 
                            categories=['Pas du tout', 'Un peu' 
                                        , 'Beaucoup', 'Sans opinion'] 
                                        , ordered=True)
```

```python
# Application de l'ACM
acm_items22 = prince.MCA(n_components = 2)

acm_items22 = acm_items22.fit(data[Items22])
```

```python
ax = acm_items22.plot(X=data[Items22])
ax
```

```python
principal_data22= acm_items22.row_coordinates(data)
```

```python
principal_data22.rename(columns = {0:'Compoenent 0', 1 : 'Componenet 1'}, inplace = True)

principal_data22
```

```python
acmdata22 = pd.concat([principal_data22, data[Items22]], axis = 1)
acmdata22
```

```python
# representation selon certaines conditions (les personnes qui font beaucoup confiance)
ScientifiquesUniversitairesConfiance_beauc= acmdata22[acmdata22['ScientifiquesUniversitairesConfiance']== 'Beaucoup']
ScientifiquesIndustrieConfiance_beauc= acmdata22[acmdata22['ScientifiquesIndustrieConfiance']== 'Beaucoup']
IngenieursConfiance_beauc= acmdata22[acmdata22['IngenieursConfiance']== 'Beaucoup']


# PLOT
plt.scatter(ScientifiquesUniversitairesConfiance_beauc['Compoenent 0'], ScientifiquesUniversitairesConfiance_beauc ['Componenet 1'],color='blue', marker= '*', label= 'ScientifiquesUniversitairesConfiance' )
plt.scatter(ScientifiquesIndustrieConfiance_beauc['Compoenent 0'],ScientifiquesIndustrieConfiance_beauc['Componenet 1'],color= 'red', marker='v', label= 'ScientifiquesIndustrieConfiance')
plt.scatter(IngenieursConfiance_beauc['Compoenent 0'], IngenieursConfiance_beauc['Componenet 1'],color= 'green', marker='.', label='IngenieursConfiance')
plt.legend()
```
