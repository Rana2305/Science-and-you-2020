```python
def construire_tableau(id_col):
    effectif= data[id_col].value_counts()
    pourcentage = round(100*data[id_col].value_counts(normalize=True),2)
    tableau = pd.DataFrame([effectif, pourcentage]).T
    tableau.columns = ["Effectif", "Pourcentage (%)"]
    return tableau
```

```python
construire_tableau('SanteEffet')
```

```python
# analyse bivari√©e de l"effet du d√©veloppement scientifique sur la sant√©
# d√©finir une fonction du tableau crois√©
def cross_table(c1, c2):
    t_nombre = pd.crosstab(data[c1],data[c2], margins =True)
    t_percentage = pd.crosstab(data[c1], data[c2]).apply(lambda x: 100*x/sum(x), axis =1)
    t = t_nombre.copy()
    for i in range(0, t_percentage.shape[0]):
        for j in range(0, t_percentage.shape[1]):
            t.iloc[i,j] = str(t_nombre.iloc[i,j]) + " ("+str(round(t_percentage.iloc[i,j], 2))+"%)"
    return t
```

```python
# effet sur la sant√© et age
cross_table('SanteEffet','RECO_Age')
```

```python
# effet sur la sant√© et sexe
cross_table('SanteEffet','Sexe')
```

```python
# effet sur la sant√© et Departement
cross_table('SanteEffet','Departement')
```
